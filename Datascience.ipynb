```python
# @title  DATA SCIENCE PROJECT
import pandas as pd
df = pd.read_csv('bmw car data.csv')
df
```

```python
# @title checking shape and null values
print('Number of row is ',df.shape[0])
print('Number of column is ',df.shape[1])
df.info()
df.isnull().sum()
```

```python
# @title 1.Get the highest car model sales
df['Model'].value_counts().reset_index()
```

```python
# @title 2.Display the car only petrol
df[df['Fuel_Type']=='Petrol']
```

```python
# @title 3.check the different result in Sales_Classification
df['Sales_Classification'].value_counts().reset_index()
```

```python
# @title 4.Find the unique values in Transmission
df['Transmission'].unique()
df['Transmission'].value_counts().reset_index()
```

```python
# @title 5.Total number of M5 car sell in Year 2019
df[(df['Model']=='M5') & (df['Year']==2019)]
```

```python
# @title 6.What is the average sales volume per model
df.groupby('Model')['Sales_Volume'].mean().reset_index()
```

```python
# @title 7.What was the top-performing models overall and by year
df.groupby('Model')['Year'].sum().reset_index()
```

```python
# @title 8. Model Price according to type
df.groupby(by = 'Model')['Price_USD'].sum().reset_index()
```

```python
# @title 9. Highest Mileage_KM in accending order
df.sort_values(by = 'Mileage_KM')
```

```python
# @title 10.Find who have bought Model in according to region
df.groupby(by = 'Region')['Model'].sum().reset_index()
```

```python
# @title 11. How does fuel type or transmission affect vehicle resale price or mileage
df.groupby(by = 'Fuel_Type')['Price_USD'].sum().reset_index()
df.groupby(by = 'Transmission')['Mileage_KM'].sum().reset_index()
```

```python
# @title 12. Which region show the strongest growth in sales volume
df.groupby(by = 'Region')['Sales_Volume'].sum().reset_index()
```

```python
# @title 13. Displaying all the rows with model X6 and the Pricr_USD greater than 5ooo
df[(df['Model']=='X6') & (df['Price_USD']>5000)]
```

```python
# @title 14. Find the name of car with highest price_usd
highest_price = df['Price_USD'].max()
df[df['Price_USD'] == highest_price][['Model','Price_USD']].reset_index()
```

```python
# @title 15. Finf out all the model with engine_size less than 2 and count the number of models
a = df[df['Engine_Size_L']< 2]
print(a)
a['Engine_Size_L'].count()
```

```python
# @title 16. Find model that are from asia or africa and have blue color
df[(df['Region']=='Asia') | (df['Region']=='Africa')][df['Color']=='Blue']
```

```python
# @title 17. what is the average price_usd of model from row 1000 to 3000
df.loc[1000:3000,'Price_USD'].mean()
```

```python
# @title 18. Check the count of different category in fuel type
df['Fuel_Type'].value_counts().reset_index()
```

```python

# @title 19. Using iloc function to display the valu of mode,price_uds,sales_classification
df.iloc[:,[0,8,10]]

```

```python
# @title 20. Remove sales_volume from the whole table
df.drop(['Sales_Volume'],axis = 1)
```

```python
# @title 21. Find the top 10 expensive car model according to their price
import seaborn as sns
import matplotlib.pyplot as plt
top10 = df.sort_values(by = "Price_USD",ascending = False).head(10)
sns.barplot(x="Price_USD" , y = "Model" , data = top10)
plt.xlabel("Price_USD")
plt.ylabel("Model")
plt.title("Top 10 Expensive Car model")
plt.show()
```

```python
# @title 22. How the price_usd,Mileage_km,sales_volume related
sns.pairplot(df , vars=["Price_USD","Mileage_KM","Sales_Volume"] , hue="Sales_Classification")
plt.show()
```

```python
# @title 23. Show the relationship between price and mileage
plt.figure(figsize=(10,6))
plt.scatter(df['Price_USD'], df['Mileage_KM'], color='green', alpha=0.6)
plt.xlabel('Price_USD')
plt.ylabel('Mileage_KM')
plt.title('Relationship between Price and Mileage')
plt.show()
```

```python
# @title 24. Which region have the more lister car
state_count = df["Region"].value_counts().head(10)
plt.figure(figsize = (15,7))
plt.bar(state_count.index , state_count.values , color = "orange")
plt.xlabel("Region")
plt.ylabel("Number ob cars ")
plt.title("Number of car Listed by Region")
plt.show()
```

```python
# @title 25. Which models have the highest overall sales volume?
import seaborn as sns
import matplotlib.pyplot as plt
top_models = df.groupby('Model')['Sales_Volume'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_models.index, y=top_models.values)
```

```python
# @title 25. How does engine size or price vary across all cars
plt.figure(figsize=(10,6))
plt.hist(df['Engine_Size_L'], bins=20, color='violet', edgecolor='black')
plt.xlabel('Engine Size (L)')
plt.ylabel('Frequency')
plt.title('Distribution of Engine Sizes')
plt.show()
```

```python
# @title 26.How does price vary by engine size(scatter plot with regression line)
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Engine_Size_L', y='Price_USD', color='red')
plt.xlabel('Engine Size (L)')
plt.ylabel('Price (USD)')
plt.title('Price vs Engine Size')
plt.show()
```

```python
# @title 27. How does sales volume change across regions over tine
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='Sales_Volume', hue='Region', marker='o')
plt.xlabel('Year')
plt.ylabel('Sales Volume')
plt.title('Sales Volume by Region Over Time')
plt.show()
```

```python
# @title 28. compare average price by fuel type or transmission
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Fuel_Type', y='Price_USD', color='blue')
plt.xlabel('Fuel Type')
plt.ylabel('Price (USD)')
plt.title('Average Price by Fuel Type')
plt.show()
```

```python
# @title 29. What is the distribution of BMW sales by year from 2010 to 2024 (piechart)
plt.figure(figsize=(10,6))
year_counts = df['Year'].value_counts().sort_index()
plt.pie(year_counts, labels=year_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of BMW Sales by Year')
```

```python
# @title 30. How do engine size,price,and sales volume intract
sns.pairplot(df, vars=['Engine_Size_L', 'Price_USD', 'Sales_Volume'], hue='Sales_Classification')
plt.show()
```
